```python
import pandas as pd
df=pd.read_csv(r'C:\Users\saran_3jr101q\Downloads\student_performance.csv') #r is the raw string literal
df
```

```python
#total students
df1=df.shape[0]
df1
```

```python
#students who studied > 2 hrs
df2=df['student_id'].loc[df['study_hours_per_day']>2]
df2
```

```python
#students who passed assuming cutoof = 50%
df3=df['student_id'].loc[df['exam_score']>50]
df3
```

```python
df4=pd.merge(df2,df3)
df4
```

```python
#percentage of students who passed among those who studied > 2 hrs 
percentage=(df4.shape[0]/df2.shape[0])
print(str(percentage)+" %")
```

```python
#probability a student studied > 2hrs given they passed
#A-- event student studied greater than 2 hrs
#B-- event where student passed
#find P(A|B)== (P(B|A)*P(A))/P(B)
#P(B|A)=P(B intersection A)/P(A)
PA=df2.shape[0]/df1
PB=df3.shape[0]/df1
PBA=df4.shape[0]/df2.shape[0]
PAB=(PBA*PA)/PB
print(PAB)
```
